```python
from googletrans import Translator
import pandas as pd
from csv import writer
translator = Translator()
```

```python
# has to be run only the first time the user uses this program

dict = {
'English': [],
'Hindi': []
}
with open('words_main.txt', 'r') as words:
    for word in words:
        try:
            # stripped_word = word.strip()
            out = translator.translate(word, 'hi', 'en')
            dict['English'].append(word)
            dict['Hindi'].append(out.text)
            print(f'{word} : {out.text}')
        except:
            pass

words_df = pd.DataFrame(dict)
words_df['English'] = words_df['English'].str.replace(r'\n', '')
words_df

words_df.to_csv('words.csv',index=False)
will_be_turned_into_csv = pd.DataFrame({'English':[], 'Hindi':[]})
will_be_turned_into_csv = will_be_turned_into_csv.append(words_df, ignore_index = True)
```

```python
words_to_be_added = ['peculiar']
```

```python
df=pd.read_csv('words.csv')
for word in words_to_be_added:
    exists_count = df['English'].str.contains(word).sum()
    if not exists_count>0:
        with open("words_main.txt", "a+") as file_object:
            # Move read cursor to the start of file.
            file_object.seek(0)
            # If file is not empty then append '\n'
            data = file_object.read(100)
            if len(data) > 0 :
                file_object.write("\n")
            # Append text at the end of file
            file_object.write(word)
            print(f"Wrote {word} to the txt file")
            file_object.close()
        trans = translator.translate(word, 'hi', 'en')
        new_words = pd.DataFrame({'English': [word], 'Hindi':[trans.text]})
        print(f'{word} : {trans.text}')
        df = df.append(new_words, ignore_index=True)
        
    else:
        print(f"This word(s) has already been added : \n{new_words}")

df.to_csv('words.csv',index=False)
```
